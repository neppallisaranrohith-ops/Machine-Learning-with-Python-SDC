```python
for it in range(1,itr+1):
    y_pred = X @ w + b
    #print(y_pred)
    error = y_pred - y
    #print(error)
    #gradinets
    dw = (2.0/n)*(X.T @ error)
    db = (2.0/n)*np.sum(error)
    w = w - lr*dw
    b = b - lr*db
    loss = (1.0/n)*np.sum(error**2)
    print(loss)
```

```python
# ---- Data: 5 samples, 2 features ----
# Feature 1 (same as before)
x1 = np.array([-1, 0, 1, 2, 3], dtype=np.float32)

# Feature 2 (example; replace with YOUR second feature values, length must be 5)
x2 = np.array([0, 1, 2, 3, 4], dtype=np.float32)

# Stack into X matrix of shape (n, p) = (5, 2)
X = np.column_stack([x1, x2]).astype(np.float32)  # shape: (5,2)
```

```python
X
```

```python
# Targets
y = np.array([1, 2, 3, 4, 5], dtype=np.float32).reshape(-1, 1)  # shape: (5,1)

# ---- Hyperparameters ----
alpha = 0.05
max_iters = 10000
```

```python
# ---- Initialize parameters ----
n, p = X.shape
w = np.zeros((p, 1), dtype=np.float32)  # shape: (2,1)
b = np.float32(0.0)
```

```python
w.shape
```

```python
for it in range(1, max_iters + 1):
    # 1) Predict
    y_pred = X @ w + b                  # shape: (n,1)

    # 2) Error
    err = y_pred - y                    # shape: (n,1)

    # 3) Gradients (MSE)
    dw = (2.0 / n) * (X.T @ err)        # shape: (p,1)
    db = (2.0 / n) * np.sum(err)        # scalar

    # 4) Update
    w_new = w - alpha * dw
    b_new = b - alpha * db
    loss = (1.0 / n) * np.sum(err ** 2)
    print(f"iter={it:5d}  loss={loss:.10f}  w={w.ravel()}  b={float(b):.10f}")
    w, b = w_new, b_new

print("\nFinal parameters:")
print("w =", w.ravel())   # [w1, w2]
print("b =", float(b))

# Final predictions (sanity check)
y_pred_final = (X @ w + b).ravel()
print("\nPredictions:", y_pred_final)
print("Targets:     ", y.ravel())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Data (1 feature)
# -----------------------
X = np.array([-1, 0, 1, 2, 3, 4, 5], dtype=np.float32).reshape(-1, 1)
y = np.array([1, 2, 3, 4, 5, 6, 7], dtype=np.float32).reshape(-1, 1)

n = X.shape[0]
x_flat = X.ravel()
y_flat = y.ravel()

# -----------------------
# Hyperparameters
# -----------------------
alpha = 0.05
max_iters = 200
snapshot_every = 10   # <-- new plot every 10 iterations
max_snapshots = 25    # safety limit so you don't create hundreds of figures

# -----------------------
# Initialize
# -----------------------
w = np.float32(0.0)
b = np.float32(0.0)

# Grid for drawing the fitted line
x_grid = np.linspace(float(x_flat.min()) - 1, float(x_flat.max()) + 1, 200, dtype=np.float32)

snap_count = 0

for it in range(1, max_iters + 1):
    # 1) predict
    y_pred = w * X + b                       # (n,1)

    # 2) error
    err = y_pred - y                         # (n,1)

    # 3) loss
    loss = float(np.mean(err ** 2))

    # 4) gradients
    dw = float((2.0 / n) * np.sum(err * X))  # scalar (because 1 feature)
    db = float((2.0 / n) * np.sum(err))      # scalar

    # 5) update
    w = np.float32(w - alpha * dw)
    b = np.float32(b - alpha * db)

    # ---- snapshot plot ----
    if it % snapshot_every == 0 or it == 1:
        snap_count += 1
        if snap_count > max_snapshots:
            break

        y_line = (w * x_grid + b)
        y_pred_points = (w * x_flat + b)

        fig, ax = plt.subplots(figsize=(9, 6))
        ax.scatter(x_flat, y_flat, label="Targets")
        ax.scatter(x_flat, y_pred_points, marker="x", label="Predicted")
        ax.plot(x_grid, y_line, label="Model line")

        ax.set_xlabel("X")
        ax.set_ylabel("y")
        ax.set_title(f"Snapshot @ iter={it}")

        info = (
            f"iter: {it}\n"
            f"loss (MSE): {loss:.6f}\n"
            f"w: {float(w):.6f}\n"
            f"b: {float(b):.6f}\n"
            f"pred: {np.round(y_pred_points, 3)}\n"
            f"tgt : {np.round(y_flat, 3)}"
        )
        ax.text(0.02, 0.98, info, transform=ax.transAxes, va="top")
        ax.legend()
        ax.grid(True)

        plt.show()

print("Final w:", float(w))
print("Final b:", float(b))
```
